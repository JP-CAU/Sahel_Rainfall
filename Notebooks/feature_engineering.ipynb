# Experiment with how to compute new features

```python
import pandas as pd
import numpy as np

climind = pd.read_csv("climate_indices.csv")
climind = climind.set_index(["model", "year", "month", "index"]).unstack(level=-1)[
    "value"
]

cesm = climind.loc[climind.index.get_level_values('model') == 'CESM']
foci = climind.loc[climind.index.get_level_values('model') == 'FOCI']

df_cesm = cesm["PREC_SAHEL"].reset_index(drop=True)
df_foci = foci["PREC_SAHEL"].reset_index(drop=True)
new_feat = list()
for i in range(len(df_cesm)):
    if i <= 2:
        new_feat.append(None)
    else:
        new_feat.append(np.mean([df_cesm.iloc[i - 1], df_cesm.iloc[i - 2], df_cesm.iloc[i - 3]]))
for i in range(len(df_foci)):
    if i <= 2:
        new_feat.append(None)
    else:
        new_feat.append(np.mean([df_foci.iloc[i - 1], df_foci.iloc[i - 2], df_foci.iloc[i - 3]]))
climind['last_3_years'] = new_feat
climind
```
